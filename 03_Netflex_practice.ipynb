```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta

# Set seed for reproducibility
np.random.seed(42)
random.seed(42)

# Create a Faker instance for generating fake data
faker = Faker()

# Number of virtual customers
num_customers = 10000

# Generate synthetic customer data
customer_data = {
    'Customer_ID': [f'C{i+1}' for i in range(num_customers)],
    'Name': [faker.name() for _ in range(num_customers)],
    'Age': [random.randint(18, 65) for _ in range(num_customers)],
    'Location': [faker.country() for _ in range(num_customers)],
    'Subscription_Type': [random.choice(['Basic', 'Standard', 'Premium']) for _ in range(num_customers)],
}

# Create a DataFrame for customer data
customer_df = pd.DataFrame(customer_data)

# Simulate Netflix transactions
start_date = datetime(2022, 1, 1)
end_date = datetime(2023, 12, 31)

# Generate random viewing history for each customer
viewing_history = []
for customer_id in customer_df['Customer_ID']:
    num_views = random.randint(1, 100)
    for _ in range(num_views):
        date_watched = faker.date_between(start_date=start_date, end_date=end_date)
        duration_watched = timedelta(minutes=random.randint(10, 180))
        title_watched = faker.random_element(elements=['Title1', 'Title2', 'Title3'])

        viewing_history.append({
            'Customer_ID': customer_id,
            'Date_Watched': date_watched,
            'Duration_Watched': duration_watched,
            'Title_Watched': title_watched,
        })

# Create a DataFrame for viewing history
viewing_df = pd.DataFrame(viewing_history)

# Save customer data and viewing history to CSV files
#customer_df.to_csv('customer_data.csv', index=False)
#viewing_df.to_csv('viewing_history.csv', index=False)
```
