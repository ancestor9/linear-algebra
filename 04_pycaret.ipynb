```python
# load dataset
from pycaret.datasets import get_data
diabetes = get_data('diabetes')

# init setup
from pycaret.classification import *
clf1 = setup(data = diabetes, target = 'Class variable')
```

```python
# list models
models()
```

```python
classifiers_list = list(models().index)
classifiers_list.remove('svm')
classifiers_list.remove('ridge') # SVM은 prediction score가 없음
classifiers_list
```

```python
from sklearn import metrics
```

```python
dir(metrics)
```

```python
clf = create_model('lr', fold=5)
predictions = predict_model(clf)
predictions
```

```python
results = {}

# train model & prediction
for ml in classifiers_list:
    # train model
    clf = create_model(ml, fold=3)
    # tune model
    tuned_clf = tune_model(clf, fold=3)

    predictions = predict_model(tuned_clf)
    y_test = predictions['Class variable']
    y_pred = predictions['prediction_label']
    y_score = predictions['prediction_score']
    tn, fp, fn, tp = metrics.confusion_matrix(y_test, y_pred).ravel()

    result = {
        'Best Parameters': clf.get_params(),  # You need to define 'clf' before using it
        'accuracy': metrics.accuracy_score(y_test, y_pred),
        'precision': metrics.average_precision_score(y_test, y_score),
        'Recall_macro': metrics.recall_score(y_test, y_pred, average='macro'),
        'f1-score': metrics.f1_score(y_test, y_pred),
        'Kappa': metrics.cohen_kappa_score(y_test, y_pred),  # Specify y_test, y_pred here
        'MCC': metrics.matthews_corrcoef(y_test, y_pred),
        'AUC': metrics.roc_auc_score(y_test, y_score),
        'Brier_scor': metrics.brier_score_loss(y_test, y_score),
        'TN(T0-P0)' : tn,
        'FP(T0-P1)' : fp,
        'FN(T1-P0)' : fn,
        'TP(T1-P1)' : tp,
    }
    # Update the 'results' dictionary
    results.update({ml: result})
```

```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.2f' % x)

result = pd.DataFrame(results).T
result
```
