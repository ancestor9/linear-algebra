```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# 예제 데이터 생성
X = 2 * np.random.rand(100, 1)  # 독립 변수
y = 4 + 3 * X + np.random.randn(100, 1)  # 종속 변수 (선형 관계에 노이즈 추가)

# Scipy를 사용한 선형 회귀
slope, intercept, r_value, p_value, std_err = linregress(X.flatten(), y.flatten())

# 학습된 선형 회귀 모델의 파라미터 출력
print(f'기울기 (slope): {slope}, 절편 (intercept): {intercept}')

# 새로운 데이터에 대한 예측
X_new = np.array([0, 2]) # X의 최소값과 최대값에 대한 예측
y_predict = intercept + slope * X_new
print("예측값:", y_predict)

# 학습된 선형 회귀 모델 시각화
plt.plot(X_new, y_predict, "r-")
plt.scatter(X, y)
plt.xlabel('X'); plt.ylabel('y')
plt.title('Simple Linear Regression with Scipy')
```

```python
import hashlib

def simple_hash_function(input_string):
    # Create a new hash object using SHA-256 algorithm
    hash_object = hashlib.sha256()

    # Update the hash object with the input string encoded as bytes
    hash_object.update(input_string.encode('utf-8'))

    # Get the hexadecimal representation of the hash
    hash_code = hash_object.hexdigest()

    return hash_code

# Example usage
input_data = "Hello, World!"
hash_result = simple_hash_function(input_data)
print(f"Input: {input_data}\nHash Code: {hash_result}")
```

```python

# Define the log loss function
def log_loss(true_labels, predicted_probabilities):
    epsilon = 1e-15  # Small constant to prevent log(0)
    predicted_probabilities = np.clip(predicted_probabilities, epsilon, 1 - epsilon)
    return -np.mean(true_labels * np.log(predicted_probabilities) + (1 - true_labels) * np.log(1 - predicted_probabilities))

# Generate a range of predicted probabilities
predicted_probs = np.linspace(0.01, 0.99, 100)

# Create a plot for binary classification
true_label_binary = 1  # Binary classification with true label 1
log_loss_binary = [log_loss(true_label_binary, p) for p in predicted_probs]

plt.figure(figsize=(8, 6))
plt.plot(predicted_probs, log_loss_binary, label='Binary Classification (True Label=1)')
plt.title('Log Loss Function for Binary Classification')
plt.xlabel('Predicted Probability')
plt.ylabel('Log Loss')
plt.legend()
plt.grid(True)
```

```python

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate a range of values for z
z_values = np.linspace(-7, 7, 200)

# Apply the sigmoid function to the range of z values
sigmoid_values = sigmoid(z_values)

# Plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(z_values, sigmoid_values, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('z')
plt.ylabel('sigmoid(z)')
plt.legend()
plt.grid(True)
```

```python
# Tokenize the text into words
words = word_tokenize(hamlet_text)

# Convert to lowercase to treat words case-insensitively
words = [word.lower() for word in words]

# Filter out non-alphabetic words
words = [word for word in words if word.isalpha()]

# Calculate word frequencies
word_frequencies = Counter(words)

# Get the 10 most common words
top_10_words = word_frequencies.most_common(10)

# Print the result
print("Top 10 most frequently occurring words in Hamlet:")
for word, frequency in top_10_words:
    print(f"{word}: {frequency}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define two 2D vectors
x = np.array([2, 1]); y = np.array([-1, 3])

# Define scalar coefficients for linear combinations
coefficients = np.linspace(-2, 2, 10)

# Calculate linear combinations
combinations = np.array([c*x + d*y for c in coefficients for d in coefficients])

# Plot the vectors and linear combinations
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')

for i in range(len(combinations)):
    plt.quiver(0, 0, combinations[i, 0], combinations[i, 1], angles='xy', scale_units='xy', scale=1, color='g', alpha=0.5)

plt.xlim(-8, 8); plt.ylim(-8, 8)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.title('Linear Combinations of 2D Vectors')
```

```python
# Define two 2D vectors
vector_a = np.array([1, 2]); vector_b = np.array([4, 3])

# Calculate dot product
dot_product = np.dot(vector_a, vector_b)

print(dot_product)

# Plot the vectors
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A')
plt.quiver(0, 0, vector_b[0], vector_b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector B')

# Add annotations for vectors
plt.text(vector_a[0], vector_a[1], 'A', fontsize=12, color='r', ha='right', va='bottom')
plt.text(vector_b[0], vector_b[1], 'B', fontsize=12, color='b', ha='right', va='bottom')

# Set axis limits with adjusted length
plt.xlim(-1, 8); plt.ylim(-1, 5)

# Plot the dot product as a line
plt.quiver(0, 0, dot_product * vector_b[0], dot_product * vector_b[1],
           angles='xy', scale_units='xy', scale=1, color='g', label='Dot Product')

# Add annotation for dot product
plt.text(dot_product * vector_b[0], dot_product * vector_b[1], 'A · B', fontsize=12, color='g', ha='right', va='bottom')

# Plot the length of the dot product as a vector
plt.quiver(0, 0, dot_product, 0, angles='xy', scale_units='xy', scale=1, color='purple', label='Length of Dot Product')

# Add annotation for the length of dot product
plt.text(dot_product, 0, f'|A · B| = {abs(dot_product):.2f}', fontsize=12, color='purple', ha='right', va='bottom')

# Add a legend
plt.legend()

# Add grid for reference
plt.grid(True, linestyle='--', alpha=0.5)
```

```python
A = np.array([3,4])
B = np.array([4,1])
# Calculate the dot product
dot_product = np.dot(A, B) # A@B

# Calculate the magnitudes of the vectors
magnitude_A = np.linalg.norm(A) ; magnitude_B = np.linalg.norm(B)
# Calculate the cosine of the angle
cosine_angle = dot_product / (magnitude_A * magnitude_B)
# Display the result
print(f"The cosine between A and B is: {cosine_angle}")

```

```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_l1_norm(data, norm):
    # Calculate L1 norms for each data point
    l1_norms = np.linalg.norm(data, ord=norm, axis=1)

    # Scale data based on L1 norms
    scaled_data = data / l1_norms[:, np.newaxis]

    # Plotting
    plt.scatter(data[:, 0], data[:, 1], label='Original Data', alpha=0.5)
    plt.scatter(scaled_data[:, 0], scaled_data[:, 1],
                label=f'Scaled with with Norm_{norm}', alpha=0.5)

    plt.axhline(0); plt.axvline(0)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)

    plt.title(f'Visualization of 2D Data Scaled with Norm_{norm}')
    plt.xlabel('X-axis'); plt.ylabel('Y-axis')
    plt.legend()
    plt.show()

# Generate random 2D data
random_data = np.random.rand(1000, 2) * 4 - 2  # Values between -2 and 2 for both dimensions

# Visualize the data scaled with L1 norm
visualize_l1_norm(random_data, norm = 2) # Eucledean Norm (L2 norm)
```
